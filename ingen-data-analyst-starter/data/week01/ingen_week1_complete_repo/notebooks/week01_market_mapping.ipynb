# Week 1 Market Mapping
Loads product profiles, competitor landscape, and data dictionary. Produces basic summary tables.


```python
import pandas as pd
from pathlib import Path
base = Path('..')
products = pd.read_excel(base/'data/week01/product_profiles.xlsx')
competitors = pd.read_excel(base/'data/week01/competitor_landscape.xlsx')
data_dictionary = pd.read_excel(base/'data/week01/data_dictionary.xlsx')
products.head()
```

```python
competitors.groupby('vertical').agg(companies=('company','count')).reset_index()
```

```python
summary = competitors.groupby('vertical')['company'].count().sort_values(ascending=False)
ax = summary.plot(kind='bar', title='Competitor Count by Vertical', figsize=(8,4))
ax.set_xlabel('Vertical')
ax.set_ylabel('Competitor count')
```
