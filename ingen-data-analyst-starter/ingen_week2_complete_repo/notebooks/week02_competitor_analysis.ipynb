# Week 2 - Competitive Intelligence Analysis

This notebook loads the Week 2 peer profiles and produces simple summary views used in the dossier.


```python
import pandas as pd
peers = pd.read_excel('../data/week02/competitor_profiles.xlsx', sheet_name='peer_profiles')
hiring = pd.read_excel('../data/week02/hiring_signals.xlsx', sheet_name='roles_snapshot')
peers.head()
```

```python
vertical_counts = peers.groupby('vertical')['company'].count().reset_index(name='peer_count')
vertical_counts
```

```python
hiring_summary = hiring.groupby('vertical').agg(total_roles=('total_open_roles_reviewed','sum'), ai_ml_roles=('ai_ml_roles','sum')).reset_index()
hiring_summary['ai_ml_share'] = hiring_summary['ai_ml_roles'] / hiring_summary['total_roles']
hiring_summary.sort_values('ai_ml_share', ascending=False)
```
